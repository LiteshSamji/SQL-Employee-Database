```python
from sqlalchemy import create_engine
# Create a database connection
engine = create_engine(f"postgresql://{user}:{password}@localhost:5432/Employee_db")
connection = engine.connect()
```

```python
#Create Dataframe to employee data
employees_df = pd.read_sql("select * from employees", connection)
#Replace Null Values
employees_df.fillna("NaN")
```

```python
#Create dataframe to hold salary data 
salary_df = pd.read_sql("SELECT * FROM Salaries", connection)
salary_df
```

```python
#Dataframe for title
titles_df = pd.read_sql("SELECT * FROM Titles", connection)
titles_df = titles_df.rename(columns = {"title_id": "emp_title_id"})
titles_df
```

```python
# Bonus question 1: Histogram to visualize the most common salary ranges for employees
plt.hist(salary_df['salary'], label="Salary")
plt.axvline(salary_df['salary'].mean(), color='Blue', linestyle='dashed', linewidth=2, label="Salary Mean")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.legend() 
plt.title("Common Salary Ranges")
plt.savefig("Common Salary Ranges.png")
```

```python
# Bonus quesion 2: Create a bar chart of average salary by title

#### Create a merged dataframe of employees and salaries
merge_df = pd.merge(employees_df, salary_df, on = "emp_no")
merge_df.head()
```

```python
#Calculate average salaries for all employees
#Merge employees and Salary Data Frame

merge_df = pd.merge(employees_df, salary_df, on = "emp_no")
merge_df.head()
```

```python
#Calculate the Average mean salary
avg_salary_by_title = merge_df.groupby(["emp_title_id"])["salary"].mean()
avg_salary_by_title
```

```python
#Merge the Titles data frame with the average salary data
salary_by_title_df = pd.merge(titles_df, avg_salary_by_title, on = "emp_title_id")
salary_by_title_df
```

```python
# Diplay the bar chart of average salary by title
x_axis = np.arange(len(salary_by_title_df))
tick_locations = [value for value in x_axis]
salary_by_title_df = salary_by_title_df.sort_values("salary", ascending = False)
plt.bar(x_axis, salary_by_title_df["salary"], alpha = 0.9, align="center")
plt.xticks(tick_locations, salary_by_title_df["title"], rotation="45", horizontalalignment = "right")
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, 60000)
plt.title("Average Salary by Title", fontsize = 14)
plt.xlabel("Title")
plt.ylabel("Average Salary")
plt.show()
```
